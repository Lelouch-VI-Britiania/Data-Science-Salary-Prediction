```python
df = pd.read_csv('/content/drive/MyDrive/glassdoor/SalaryData_cleaned.csv')
df
```

```python
def describe_statistics(df):
    stats = []

    for col in df.select_dtypes(include=np.number).columns:
        col_data = df[col].dropna()
        percentiles = np.percentile(col_data, [25, 50, 75])
        stats.append({
            'Variable': col,
            'Frequency': '-',  # Optional for numeric
            'Percentile (25th, 50th, 75th)': f"{percentiles[0]:.2f}, {percentiles[1]:.2f}, {percentiles[2]:.2f}",
            'Data Completeness': f"{100 * col_data.count() / len(df):.1f}%",
            'Mean': round(col_data.mean(), 2),
            'Median': round(col_data.median(), 2),
            'Mode': col_data.mode().values[0],
            'Standard Deviation': round(col_data.std(), 2),
            'Variance': round(col_data.var(), 2),
            'Skewness': round(col_data.skew(), 2),
            'Kurtosis': round(col_data.kurtosis(), 2)
        })

    for col in df.select_dtypes(exclude=np.number).columns:
        freq = df[col].value_counts(normalize=True).head(3).to_dict()
        stats.append({
            'Variable': col,
            'Frequency': ', '.join([f"{k}: {v:.1%}" for k, v in freq.items()]),
            'Percentile (25th, 50th, 75th)': '-',
            'Data Completeness': f"{100 * df[col].count() / len(df):.1f}%",
            'Mean': '-',
            'Median': '-',
            'Mode': df[col].mode().values[0],
            'Standard Deviation': '-',
            'Variance': '-',
            'Skewness': '-',
            'Kurtosis': '-'
        })

    return pd.DataFrame(stats)
```

```python
table2 = describe_statistics(df)
table2
```

```python
with pd.ExcelWriter("salary_table_statistics.xlsx") as writer:
    table2.to_excel(writer, sheet_name="Data Statistics", index=False)
```
